### Estandarizando variables

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(predictores)
```

```python
predictores_prepared = scaler.transform(predictores)
```

Separando variable respuesta del dataset

```python
y = df["aux"].values.reshape(-1,1).ravel()
```

## Modelo con K = 4

Creación del modelo

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_KNN_4 = KNeighborsClassifier(n_neighbors = 4)
```

Ajuste del modelo

```python
model_KNN_4.fit(predictores_prepared, y)
```

Clasificación obtenida

```python
# valores de x1
min_x1 = min(predictores_prepared[:,0])
max_x1 = max(predictores_prepared[:,0])
x1_values = np.linspace(min_x1, max_x1, 101)
# valores de x2
min_x2 = min(predictores_prepared[:,1])
max_x2 = max(predictores_prepared[:,1])
x2_values = np.linspace(min_x2, max_x2, 101)

x1_grid, x2_grid = np.meshgrid(x1_values, x2_values)
x1_grid = x1_grid.flatten()
x2_grid = x2_grid.flatten()
x_grid = pd.DataFrame({'x1':x1_grid, 'x2':x2_grid})

# tipos----------------------------------------------------------
pred_4 = model_KNN_4.predict(x_grid)
regiones_4 = x_grid.copy()
# pegar los tipos predichos a la rejilla-------------------------
regiones_4["clase"] = pred_4
regiones_4["clase"] = regiones_4["clase"].astype(object)
# ver cuantos hay de cada en la rejilla
print(regiones_4["clase"].value_counts())
```

Visualización de la clasificación obtenida

```python
(
    ggplot(regiones_4, aes(
    x = "x1",
    y = "x2",
    fill = "clase")) +
    geom_point(size = 1) +
    ylab("X2") +
    xlab("X1") +
    theme_bw() +
    theme(legend_position = "right",
    subplots_adjust={'right': 0.8})
)
```

## Determinación de K óptimo

Valores de K que vamos a probar

```python
k_range = range(1, 46)
```

Inicializando el vector de puntuaciones

```python
k_scores = []
```

Creación de rutina para extraer las las puntuaciones

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    
    scores = cross_val_score(knn, 
                             predictores_prepared, 
                             y,
                             cv = 15, 
                             scoring = "accuracy")

    k_scores.append(scores.mean())
```

Variación de accuarcy con K

```python
print(k_scores)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(k_range, k_scores)
plt.xlabel('Valor de K para KNN')
plt.ylabel('Accuracy de VC')
plt.show()
```

```python
print("El valor de K que maximiza la accuracy es",
np.argmax(k_scores) + 1)
```
