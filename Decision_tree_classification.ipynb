## Preparación de datos

```python
X = df.drop(columns = ['tipo'])
y = df['tipo']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    train_size = 0.8,
                                                    random_state = 42,
                                                    shuffle = True
                                                   )
```

## Modelo

### Modelo por defecto

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.metrics import accuracy_score
```

```python
tree_clf = DecisionTreeClassifier(random_state = 3);
tree_clf.fit(X_train, y_train);
y_pred_tree = tree_clf.predict(X_test);
print("El valor de accuracy es:", accuracy_score(y_test, y_pred_tree))
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
import seaborn as sns
```

```python
fig, ax = plt.subplots(figsize=(12, 10))

print(f"Profundidad del árbol: {tree_clf.get_depth()}")
print(f"Número de nodos terminales: {tree_clf.get_n_leaves()}")

plot = plot_tree(
            decision_tree = tree_clf,
            feature_names = X_train.columns,
            class_names   = 'MEDV',
            filled        = True,
            impurity      = False,
            fontsize      = 10,
            precision     = 2,
            ax            = ax
       )
```

```python
result = {'predictor': X_train.columns,'importancia': tree_clf.feature_importances_}
predictores = pd.DataFrame(result)
print("Importancia de los predictores en el modelo")
predictores.sort_values('importancia', ascending=False)
```

### Nivel de profundidad máxima igual a 4

```python
tree_clf2 = DecisionTreeClassifier(random_state = 3,max_depth = 4);
tree_clf2.fit(X_train, y_train);
y_pred_tree2 = tree_clf2.predict(X_test);
print("El valor de accuracy es:", accuracy_score(y_test, y_pred_tree2))
```

```python
fig, ax = plt.subplots(figsize=(12, 10))

print(f"Profundidad del árbol: {tree_clf2.get_depth()}")
print(f"Número de nodos terminales: {tree_clf2.get_n_leaves()}")

plot = plot_tree(
            decision_tree = tree_clf2,
            feature_names = X_train.columns,
            class_names   = 'MEDV',
            filled        = True,
            impurity      = False,
            fontsize      = 10,
            precision     = 2,
            ax            = ax
       )
```

```python
result = {'predictor': X_train.columns,'importancia': tree_clf2.feature_importances_}
predictores = pd.DataFrame(result)
print("Importancia de los predictores en el modelo")
predictores.sort_values('importancia', ascending=False)
```
